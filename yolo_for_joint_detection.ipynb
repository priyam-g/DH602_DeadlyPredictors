```python
image_path = '/content/drive/MyDrive/DH602_Project/YOLO/DH602_YOLO2/Splitted_images/images'
```

```python
import os

from ultralytics import YOLO

model = YOLO(model_path)
```

```python
images_file_path = sorted(os.listdir(image_path))
```

```python
import cv2 as cv
import numpy as np
images = []
img_paths = []
for image_file in images_file_path:
  img = cv.imread(os.path.join(image_path, image_file))
  img_path = os.path.join(image_path, image_file)
  img_paths.append(img_path)
  # images.append(img)
  # img = cv.imread(os)
  # img = cv.resize(img, (512, 512))
  images.append(img)

# images = np.array(images)
```

```python
print(img.shape)
```

```python
def is_any_dimension_zero(arr):
    return 0 in arr.shape
```

```python
import torch
import matplotlib.pyplot as plt

count = 0
problem = []
for i in range(len(images)-1):
  img_full = images[i]
  # plt.imshow(img_full)
  # plt.show()


  results = model.predict(img_paths[i], imgsz=512, save=True, conf=0.5)
  for result in results:
    boxes = result.boxes


  bboxes = boxes.xyxy

  for bbox in bboxes:
    bbox = torch.round(bbox).to(torch.int)
    cropped_img = img_full[bbox[1] - 20 :bbox[3] + 20, bbox[0] - 20 :bbox[2] + 20]
    print(cropped_img.shape)
    # plt.imshow(cropped_img)
    # plt.show()
    if is_any_dimension_zero(cropped_img):
      problem.append(count)
      count +=1
      continue
    cv.imwrite(f'/content/drive/MyDrive/DH602_Project/YOLO/DH602_YOLO2/Cropped_images/{count}.jpg', cropped_img)
    count +=1
  # break
```

```python
import random

test_img = random.choice(img_paths)

results = model.predict(test_img, save=True, imgsz=512, conf=0.5)
```
