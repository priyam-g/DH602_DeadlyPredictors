```python
import pandas as pd
import cv2 as cv
import os
import numpy as np
```

```python
image_folder_path = '/Users/sandipandas/Desktop/AIRINIIReumHands/jpeg'
csv_path = '/Users/sandipandas/Desktop/AIRINIIReumHands/bboxes.csv'

images = sorted(os.listdir(image_folder_path))
df = pd.read_csv(csv_path)
print(len(images))
df.head()
```

```python
test_list = [7, 411, 477, 209, 211,358, 8, 411, 472, 388, 299]
for test in test_list:
    df = df.drop(df[df['patient_id'] == test].index)
df.head()
```

```python
joint_labels = {'DIP':0, 'CMC':1, 'wrist':2, 'RC':3, 'ulna':4, 'PIP':5, 'MCP':6}
df['joint_labels'] = df['joint'].map(joint_labels)
df.head()
```

```python
target_label_folder = '/Users/sandipandas/Desktop/Data/Train/labels'
target_image_folder = '/Users/sandipandas/Desktop/Data/Train/images'
```

```python
selected = ['joint_labels', 'xcenter_norm', 'ycenter_norm', 'dx_norm', 'dy_norm']
for image_name in images:
    img = cv.imread(os.path.join(image_folder_path, image_name))
    idx = np.int64(image_name.split('.')[0])
    t = df[df['patient_id'] == idx]
    t['xcenter_norm'] = t['xcenter']/img.shape[1]
    t['ycenter_norm'] = t['ycenter']/img.shape[0]
    t['dx_norm'] = t['dx']/img.shape[1]
    t['dy_norm'] = t['dy']/img.shape[0]

    file_path = f'{target_label_folder}/{idx}.txt'
    t[selected].to_csv(file_path, sep='\t', index=False, header=False)
    img_path = f'{target_image_folder}/{idx}.jpg'
    cv.imwrite(img_path, img)
    # break

    
```
